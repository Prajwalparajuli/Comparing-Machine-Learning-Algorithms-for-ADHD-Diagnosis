### Merging Functional Connectivity Matrix data


```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell 
InteractiveShell.ast_node_interactivity = 'all'
```

```python
KKI = pd.read_csv("Preprocessed FC/KKI_FC.csv")
Neuro = pd.read_csv("Preprocessed FC/NeuroIMAGE_FC.csv")
NYU = pd.read_csv("Preprocessed FC/NYU_FC.csv")
OHSU = pd.read_csv("Preprocessed FC/OHSU_FC.csv")
Pek1 = pd.read_csv("Preprocessed FC/Peking_1_FC.csv")
Pek2 = pd.read_csv("Preprocessed FC/Peking_2_FC.csv")
Pek3 = pd.read_csv("Preprocessed FC/Peking_3_FC.csv")
```

```python
KKI.shape
np.isnan(KKI).sum().sum()

NYU.shape
np.isnan(NYU).sum().sum()

Neuro.shape
np.isnan(Neuro).sum().sum()

OHSU.shape
np.isnan(OHSU).sum().sum()

Pek1.shape
np.isnan(Pek1).sum().sum()

Pek2.shape
np.isnan(Pek2).sum().sum()

Pek3.shape
np.isnan(Pek3).sum().sum()
```

```python
FC_data_merged = pd.concat([KKI, Neuro, NYU, OHSU, Pek1, Pek2, Pek3])
FC_data_merged = FC_data_merged.loc[:, ~FC_data_merged.columns.str.contains('^Unnamed')]

FC_data_merged.set_index("ScanDir ID", inplace = True)

FC_data_merged
```
