#### Using Cross Validation and training the model

##### Hyperparameter Optimization with Grid Search (Support Vector Machine)

In this section, we perform **hyperparameter tuning** on a **Support Vector Machine (SVM)** classifier using **GridSearchCV** from scikit-learn. The tuning is conducted on the *Functional Connectivity + Phenotypic* dataset to identify the best combination of model parameters and then apply the optimized model to other feature sets (FC, PCA, ReHo, and fALFF).

##### What is being done?

- We use **GridSearchCV** to search over a defined range of SVM hyperparameters including:
  - Regularization strength (`C`) â€” balances margin size vs classification error.
  - Kernel type (`kernel`) â€” either `'linear'` or `'rbf'`.
  - Kernel coefficient (`gamma`) â€” controls influence of support vectors in the `'rbf'` kernel.
- An **SVM classifier** is used with class balancing enabled via `class_weight='balanced'`.
- The search is performed over all combinations using **10-fold Stratified Cross-Validation** to ensure reliable model evaluation while preserving class distribution.

##### Why is this important?

- SVMs are well-suited for **high-dimensional** data like fMRI-derived features, but their performance heavily depends on hyperparameter settings.
- The right combination of `C` and `gamma` allows the model to separate classes effectively without overfitting.
- Using **GridSearchCV** ensures thorough evaluation across combinations to select the most effective hyperparameters.
- **StratifiedKFold** ensures each fold has the same proportion of ADHD and non-ADHD subjects, which is critical for balanced evaluation.

##### Expected Outcome

- Identification of the **best-performing hyperparameters** for the SVM classifier.
- Improved **f1-score** and generalization on unseen ADHD data by tuning model complexity and decision boundaries.
- Deployment of the optimized SVM model across all feature types to determine which representation (FC, PCA, ReHo, fALFF) yields the best diagnostic performance.


```python
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, f1_score
import numpy as np

# Cross-validation setup
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=45)

# Define scoring metrics
scorings = {
    'accuracy': 'accuracy',
    'precision': 'precision_weighted',
    'recall': 'recall_weighted',
    'f1': 'f1_weighted'
}

# Define hyperparameter grid for SVM
param_grid_svm = {
     'C': [0.1, 1, 10, 100],
     'gamma': ['scale', 0.01, 0.001],
     'kernel': ['rbf'],
    'shrinking': [True, False],
     'tol': [1e-4, 1e-3, 1e-2],
    'max_iter': [-1],
    'class_weight': ['balanced']
}

# Container for storing SVM results
results_svm = []
```

#### Functional Connectivity

```python
# Step 1: Prepare X and y
X_FC = FC_data.drop(columns="DX")
y_FC = FC_data["DX"]

# Step 2: Train-test split
X_train_FC, X_test_FC, y_train_FC, y_test_FC = train_test_split(
    X_FC, y_FC, test_size=0.2, stratify=y_FC, random_state=45
)

# Step 3: Feature scaling
scaler = StandardScaler()
X_trs_FC = scaler.fit_transform(X_train_FC)
X_tes_FC = scaler.transform(X_test_FC)

# Step 4: Initialize base SVM model
base_model = SVC(class_weight='balanced', probability=True)

# Step 5: Grid Search CV
grid_search = GridSearchCV(
    estimator=base_model,
    param_grid=param_grid_svm,
    scoring=scorings,
    refit='f1',
    cv=cv,
    verbose=2,
    n_jobs=-1
)

# Step 6: Fit and evaluate
grid_search.fit(X_trs_FC, y_train_FC)
y_pred_FC = grid_search.best_estimator_.predict(X_tes_FC)

print("SVM - FC Only")
print("Best Params:", grid_search.best_params_)
print(f"Best CV F1 Score: {grid_search.best_score_:.4f}")
print(classification_report(y_test_FC, y_pred_FC, digits=4))

# Step 7: Save results
results_svm.append({
    'Dataset': 'SVM - FC Only',
    'Best Params': grid_search.best_params_,
    'CV F1': grid_search.best_score_,
    'Test F1': f1_score(y_test_FC, y_pred_FC, average='weighted')
})
```

### Another approach by combining the features into one feature 

```python
# Get common subject IDs as a list (not a set)
common_subjects = list(set(PCA_pheno.index) & set(ReHo_pheno.index) & set(fALFF_pheno.index))

# Subset all data to common subjects
Reho_df = ReHo_pheno.loc[common_subjects]
falff_df = fALFF_pheno.loc[common_subjects]
pca_df = PCA_pheno.loc[common_subjects]

# Merge the features (excluding duplicate DX columns)
merged_df = pd.concat([
    Reho_df.drop(columns=['DX']),
    pca_df.drop(columns=['DX']),
    falff_df.drop(columns=['DX']), 
], axis=1)

# Add DX back from FC_pheno_data
merged_df['DX'] = Reho_df['DX']

print(" Combined dataset shape:", merged_df.shape)
merged_df.head()
```

```python



X_merged = merged_df.drop(columns="DX")
y_merged = merged_df["DX"]

X_train_merged, X_test_merged, y_train_merged, y_test_merged = train_test_split(
    X_merged, y_merged, test_size=0.2, stratify=y_merged, random_state=45
)

scaler = StandardScaler()
X_trs_merged = scaler.fit_transform(X_train_merged)
X_tes_merged = scaler.transform(X_test_merged)


# Step 4: Initialize base SVM model
base_model = SVC(class_weight='balanced', probability=True)

# Step 5: Grid Search CV
grid_search = GridSearchCV(
    estimator=base_model,
    param_grid=param_grid_svm,
    scoring='f1_weighted',
    refit='f1_weighted',
    cv=cv,
    verbose=2,
    n_jobs=-1
)

grid_search.fit(X_trs_merged, y_train_merged)
y_pred_merged = grid_search.best_estimator_.predict(X_tes_merged)

print("SVM - fALFF + Phenotype")
print("Best Params:", grid_search.best_params_)
print(f"Best CV F1 Score: {grid_search.best_score_:.4f}")
print(classification_report(y_test_merged, y_pred_merged, digits=4))

results_svm.append({
    'Dataset': 'SVM - PCA + PeHo + fALFF + Phenotype',
    'Best Params': grid_search.best_params_,
    'CV F1': grid_search.best_score_,
    'Test F1': f1_score(y_test_merged, y_pred_merged, average='weighted')
})
```
