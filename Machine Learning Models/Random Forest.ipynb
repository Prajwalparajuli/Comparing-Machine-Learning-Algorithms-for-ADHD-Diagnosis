#### Using Cross Validation and training the model

#####  Random Forest Hyperparameter Optimization with Randomized Search

In this section, we perform **hyperparameter tuning** on a Random Forest classifier using **RandomizedSearchCV** from scikit-learn. The tuning is conducted on the *Functional Connectivity + Phenotypic* dataset to identify the best combination of model parameters and then apply the optimized model to other feature sets (FC, PCA, ReHo, and fALFF).

##### What is being done?

- We use **RandomizedSearchCV** to search over a defined range of hyperparameters including:
  - Number of trees (`n_estimators`)
  - Tree depth (`max_depth`)
  - Splitting criteria (`min_samples_split`, `min_samples_leaf`)
  - Feature selection per split (`max_features`)
  - Sampling strategy (`bootstrap`)
- A **Random Forest Classifier** is used as the base model with class balancing enabled.
- The search is performed over 25 random combinations, using **10-fold Stratified Cross-Validation** to ensure reliable model evaluation while preserving class distribution.

##### Why is this important?

- Resting-state fMRI data is high-dimensional and heterogeneous. Random Forests perform well, but **default settings are not guaranteed to be optimal**.
- **RandomizedSearchCV** is computationally more efficient than Grid Search and still provides strong results in high-dimensional feature spaces.
- By first tuning on the FC + Phenotype dataset, we create a **generalized model** that can be evaluated across all feature types for comparison.

##### Expected Outcome

- Identification of the **best-performing hyperparameters** for the Random Forest classifier.
- Improved **f1-score** and overall classification performance.
- Use of the optimized model across all feature sets to determine which feature representation (FC, PCA, ReHo, or fALFF) provides the **most predictive power for ADHD classification**.


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, f1_score
import numpy as np

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=45)

scorings = {
    'accuracy': 'accuracy',
    'precision': 'precision_weighted',
    'recall': 'recall_weighted',
    'f1': 'f1_weighted'
}

param_dist = {
    'n_estimators': [100, 300, 500, 800],
    'max_depth': [None, 10, 20, 40],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],
    'bootstrap': [True, False]
}

results = []  # to store performance for all datasets
```

#### Functional Connectivity

```python
# Step 1: Prepare X and y
X_FC = FC_data.drop(columns="DX")
y_FC = FC_data["DX"]

# Step 2: Train-test split
X_train_FC, X_test_FC, y_train_FC, y_test_FC = train_test_split(
    X_FC, y_FC, test_size=0.2, stratify=y_FC, random_state=45
)

# Step 3: Feature scaling
scaler = StandardScaler()
X_trs_FC = scaler.fit_transform(X_train_FC)
X_tes_FC = scaler.transform(X_test_FC)

# Step 4: Initialize base model
base_model = RandomForestClassifier(class_weight='balanced', random_state=45)

# Step 5: Randomized Search CV
random_search = RandomizedSearchCV(
    estimator=base_model,
    param_distributions=param_dist,
    n_iter=25,
    scoring=scorings,
    refit='f1',
    cv=cv,
    verbose=2,
    n_jobs=-1,
    random_state=45
)

# Step 6: Fit and evaluate
random_search.fit(X_trs_FC, y_train_FC)
y_pred_FC = random_search.best_estimator_.predict(X_tes_FC)

print("Random Forest - FC Only")
print("Best Params:", random_search.best_params_)
print(f"Best CV F1 Score: {random_search.best_score_:.4f}")
print(classification_report(y_test_FC, y_pred_FC, digits=4))

# Step 7: Save results
results.append({
    'Dataset': 'RF - FC Only',
    'Best Params': random_search.best_params_,
    'CV F1': random_search.best_score_,
    'Test F1': f1_score(y_test_FC, y_pred_FC, average='weighted')
})
```

### Another approach by combining the features into one feature 

```python
# Get common subject IDs as a list (not a set)
common_subjects = list(set(PCA_pheno.index) & set(ReHo_pheno.index) & set(fALFF_pheno.index))

# Subset all data to common subjects
Reho_df = ReHo_pheno.loc[common_subjects]
falff_df = fALFF_pheno.loc[common_subjects]
pca_df = PCA_pheno.loc[common_subjects]

# Merge the features (excluding duplicate DX columns)
merged_df = pd.concat([
    Reho_df.drop(columns=['DX']),
    pca_df.drop(columns=['DX']),
    falff_df.drop(columns=['DX']), 
], axis=1)

# Add DX back from FC_pheno_data
merged_df['DX'] = Reho_df['DX']

print(" Combined dataset shape:", merged_df.shape)
merged_df.head()
```

```python
# Step 1: Prepare X and y
X_combined = merged_df.drop(columns="DX")
y_combined = merged_df["DX"]

# Step 2: Train-test split
X_train_comb, X_test_comb, y_train_comb, y_test_comb = train_test_split(
    X_combined, y_combined, test_size=0.2, stratify=y_combined, random_state=45
)

# Step 3: Feature scaling
scaler = StandardScaler()
X_trs_comb = scaler.fit_transform(X_train_comb)
X_tes_comb = scaler.transform(X_test_comb)

# Step 4: Initialize base model
base_model = RandomForestClassifier(class_weight='balanced', random_state=45)

# Step 5: Randomized Search CV
random_search = RandomizedSearchCV(
    estimator=base_model,
    param_distributions=param_dist,
    n_iter=25,
    scoring=scorings,
    refit='f1',
    cv=cv,
    verbose=2,
    n_jobs=-1,
    random_state=45
)

# Step 6: Fit and evaluate
random_search.fit(X_trs_comb, y_train_comb)
y_pred_comb = random_search.best_estimator_.predict(X_tes_comb)

print("Random Forest - Combined (PCA + ReHo + fALFF + Pheno)")
print("Best Params:", random_search.best_params_)
print(f"Best CV F1 Score: {random_search.best_score_:.4f}")
print(classification_report(y_test_comb, y_pred_comb, digits=4))

# Step 7: Save results
results.append({
    'Dataset': 'RF - Combined (All)',
    'Best Params': random_search.best_params_,
    'CV F1': random_search.best_score_,
    'Test F1': f1_score(y_test_comb, y_pred_comb, average='weighted')
})
```
